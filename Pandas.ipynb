## Metodo groupby
Este metodo agrupa los registros de una columna o columnas para que podamos realizar otros metodos sobre este conjunto. pueden ser sumas, promedios , etc. estos grupos los podemos asignar o guardar en variables para su posterior uso en graficos o otras operaciones.

```python
df.groupby("Fruta")["Cantidad"].sum(numeric_only=True).sort_values(ascending=True)
```

##Cambiar nombre a las columnas
En caso de que tratemos los datos habran casos en los que el nombre de la columna no tendra nada que ver con los registros , por eso debemos cambiarle el nombre asi cuidamos la integridad de los datos.

```python
# Guardando el DataFrame en una variable
df_ejemplo = df['Tipo'].value_counts(normalize=True).to_frame().sort_values('Tipo')

# Cambiando el nombre de la columna "Tipo" a "Porcentajes"
df_ejemplo.rename(columns={'Tipo': 'Porcentajes'}, inplace=True)

# Visualizando el DataFrame
df_ejemplo
```

##Tratar datos incosistentes

```python
#Primero se debe hacer la busqueda de los indices de las filas que queremos borrar
#lo hacemos con una query y rescatamos sus indices
index_to_delete = df.query("condicion a evaluar").index
#Luego utilizamos drop() para eliminar esas filas o registros
df.drop(index_to_delete,axis = 0, inplace=True)
```
